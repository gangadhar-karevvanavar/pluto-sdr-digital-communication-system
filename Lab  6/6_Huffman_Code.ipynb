## 2. Draw Huffman Tree

- Start from the least frequent items and put these into circles, starting at the bottom right of your paper.

- Take the two items furthest to the right, pair them together, joining the branches by summing the frequencies together.

- Continue to do so until you have matched the pairs you can, leaving any spares at the moment.

- Start at the root node (In this example root node is ‘20’), put 0 next to each lefthand branch, and 1 against each right-hand branch

## 3. Read Huffman Tree to find codes

-  Start from the root node, find the path to each character (called leaf node), identify the 1 and 0 branches you have to use to get to each one.

## 5. How to Decode from Coded bits ?-  Decoding Huffman-encoded Data

- The decoding procedure is deceptively simple. To avoid ambiguity, Huffman encoding is a prefix-free encoding technique.

- No codeword appears as a prefix of any other codeword.

- To decode the encoded string, follow the zeros and ones to a leaf and return the character there.
